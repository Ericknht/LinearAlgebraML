```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')

# Cargamos la imagen
imagen = Image.open("./Imagenes/imagen_ejemplo_frida_bredesen.jpg")
imagen2 = Image.open("./Imagenes/igni.jpeg")

# Mostrar la imagen
plt.imshow(imagen)
plt.imshow(imagen2)
```

```python
# Transformamos la imagen en escala de grises
imagen_gr = imagen2.convert('LA')
print(imagen_gr)
```

```python
plt.imshow(imagen_gr)
```

```python
imagen_mat = np.array(list(imagen_gr.getdata(band=0)), float)
print(imagen_mat)
```

```python
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
print(imagen_mat)
```

```python
print(imagen_mat.shape)
```

```python
plt.imshow(imagen_mat)
```

```python
plt.imshow(imagen_mat, cmap='gray')
```

```python
imagen_mat_2 = imagen_mat / 100
print(imagen_mat_2)
```

```python
plt.imshow(imagen_mat_2)
```

```python
plt.imshow(imagen_mat_2, cmap='gray')
```

```python
print(np.max(imagen_mat_2))
print(np.max(imagen_mat))
```
